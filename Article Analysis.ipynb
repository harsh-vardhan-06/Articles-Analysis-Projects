```python
i=0
#Positive,negative score calculator
df_i['p_score']=range(0,111)
df_i['n_score']=range(0,111)
for val in df_i.corpus:
    val=nltk.word_tokenize(str(val))#Word tokenize the sentence
    n_score=0
    p_score=0
    for value in val:        
        if value in negative_words:
            n_score-=-1
        elif value in positive_words:
            p_score+=1
    df_i.p_score[i]=p_score
    df_i.n_score[i]=n_score
    i+=1
```

```python
#use to count the words in articles
def Word_count(i):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i]) 
    rp=rp.split()
    rp=[word for word in rp if not word in set (Stopwords)]
    rp=" ".join(rp)
    val=nltk.word_tokenize(str(rp))
    return len(val)    
```

```python
#use to count the sentence in articles
def sentence_count(i):
    rp=df_i['articles'][i]
    rp=rp.split()
    rp=[word for word in rp if not word in set (Stopwords)]
    rp=" ".join(rp)
    val=nltk.sent_tokenize(str(rp))
    return len(val)
```

```python
#use to count the complex words in articles
def complex_words_count(i):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i]) 
    rp=rp.split()
    rp=[word for word in rp if not word in set (Stopwords)]
    rp=" ".join(rp)
    val=nltk.word_tokenize(str(rp))
    Syllablewords=[]
    for word in val:
        if textstatistics().syllable_count(word)>2:
            Syllablewords.append(word)
    return len(Syllablewords) 
```

```python
# Word Counts After removing stopwords and words with es or ed

def SYLLABL_PER_WORD(i):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i]) 
    rp=rp.split()
    rp=[word for word in rp if not word in set (Stopwords)]
    rp=" ".join(rp)
    val=nltk.word_tokenize(str(rp))
    Syllablewords=[]
    for word in val:   #for counting syllable words
        if textstatistics().syllable_count(word)>2:
            Syllablewords.append(word)
    for value in Syllablewords: # removing words with es or ed       
        if "es" or "ed" in val:
            Syllablewords.remove(value)            
    return len(Syllablewords) 
```

```python
#polarity score
df_i['Polarity_score']=range(0,111)
for i in range(len(df_i['corpus'])):
    df_i['Polarity_score'][i]=(df_i['p_score'][i]-df_i['n_score'][i])/((df_i['p_score'][i]+df_i['n_score'][i])+0.000001)
```

```python
#Subjectivity  score
df_i['Subjectivity_score']=range(0,111)
for i in range(len(df_i['corpus'])):
    df_i['Subjectivity_score'][i]=(df_i['p_score'][i]+df_i['n_score'][i])/(Word_count(i)+0.000001)
```

```python
#Calculating average sentence length (After Removing  Stopwords)
df_i['Average sentence length']=range(0,111)
for i in range(0,111):
    df_i['Average sentence length'][i]=Word_count(i)/sentence_count(i)
```

```python
#Calculating Percentage of Complex words 
df_i['Percentage of Complex words']=range(0,111)
for i in range(0,111):
    df_i['Percentage of Complex words'][i]=complex_words_count(i)/Word_count(i)
```

```python
#Calculating FOG INDEX
df_i['FOG INDEX']=range(0,111)
for i in range(0,111):
    df_i['FOG INDEX'][i]=0.4*(df_i['Average sentence length'][i]+df_i['Percentage of Complex words'][i])
```

```python
#AVREAGE NUMBER OF WORDS PER SENTENCE (Without using Stop words)
df_i['AVG NUMBER OF WORDS PER SENTENCE']=range(0,111)
for i in range(len(df_i['AVG NUMBER OF WORDS PER SENTENCE'])):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i]) 
    df_i['AVG NUMBER OF WORDS PER SENTENCE'][i]=len(nltk.word_tokenize(str(rp)))/len(nltk.sent_tokenize(str(df_i['articles'][i])))
    
```

```python
# scalable Word Counts (After REmoving Stopwords)
df_i['COMPLEX WORD COUNT']=range(0,111)

for i in range(len(df_i['COMPLEX WORD COUNT'])):
    df_i['COMPLEX WORD COUNT'][i]=complex_words_count(i)

```

```python
# Word Counts After REmoving Stopwords with Nltk library 
from nltk.corpus import stopwords
df_i['WORD COUNT']=range(0,111)
for i in range(len(df_i['WORD COUNT'])):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i])
    rp=rp.lower()
    rp=rp.split()
    rp=[word for word in rp if not word in set(stopwords.words('english'))]
    df_i['WORD COUNT'][i]=len(rp)
```

```python
# Word Counts After removing stopwords and words with es or ed
df_i['SYLLABLE PER WORD']=range(0,111)

for i in range(len(df_i['COMPLEX WORD COUNT'])):
    df_i['SYLLABLE PER WORD'][i]=SYLLABL_PER_WORD(i)
```

```python
#Personal Pronouns
words=[ "I", "we", "my", "ours","us"]
df_i['Personal Pronouns']=range(0,111)
for i in range(0,111):
    pronounRegex = re.compile(r'I|we|my|ours|us',re.I)
    pronouns = pronounRegex.findall(df_i['articles'][i])
    df_i['Personal Pronouns'][i]=len(pronouns)
```

```python
# Average Word Length
df_i['Average Word Length']=range(0,111)
for i in range(0,111):
    rp=re.sub('[^a-zA-Z]'," ",df_i['articles'][i])
    val=nltk.word_tokenize(str(rp))
    df_i['Average Word Length'][i]=len(rp)/len(val)
```
